```python
#Neccessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Upload dataset
train = pd.read_csv("./Datasets/titanic_train.csv")
```

```python
train.head()
```

```python
#looking at null values
#False = not null
#True = is null
train.isnull()
```

```python
#now looking at the null values in a heatmap
#yellow line means null
#there are alot of null values for the Cabin and Age
sns.heatmap(train.isnull(), yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Simple histogram is survivors
#0 = did not survived
#1 = survived
sns.set_style('whitegrid')
sns.countplot(x='Survived', data=train)
```

```python
#Now faceting the graph by gender
sns.countplot(x='Survived',hue='Sex', data=train)
```

```python
#Now facting by cabin class
sns.countplot(x='Survived',hue='Pclass',data=train)
```
